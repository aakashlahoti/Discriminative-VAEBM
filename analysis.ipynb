```python
from PIL import Image
import torch
import torchvision
from torchvision import transforms
```

```python
imgs = torch.zeros(16, 3, 32, 32)
```

```python
for i in range(16):
    im = Image.open("../JEM-master-supervised/samples_all_jpg_alt/" + str(i) + ".jpg")
    convert_tensor = transforms.ToTensor()
    im = convert_tensor(im)
    imgs[i, :, :, :] = im
```

```python
torchvision.utils.save_image(imgs, "./display/JEM.jpg", nrow=4)
```
